### Boy ve Kilo Stunu Kayp Veri Doldurma
* Boy ve kilo stununda bulunan kayp veriyi etkinlik ortalamasna gre dolduracaz.

```python
# buraya kod yazacaz
# 1. Boy ve kilo stununda bulunan kayp veriyi etkinlik ortalamasna gre dolduracaz.
essiz_etkinlik = pd.unique(veri.etkinlik)
print("Esiz etkinlik says: {}".format(len(essiz_etkinlik)))
essiz_etkinlik[:10]
```

```python
# buraya kod yazacaz

# her bir etkinlii iteratif olarak dola
# etkinlik zelinde boy ve kilo ortalamalarn bul
# etkinlik zelinde boy ve kilo da kayp olan deerlere ortalama boy ve kilo deerlerini eitle

veri_gecici = veri.copy() # gerek veriyi kaybetmemek iin veri_gecici deikeni belirle
boy_kilo_liste = ["boy", "kilo"]
for e in essiz_etkinlik: # etkinlik listesi ierisinde dola
    
    # etkinlik filtresi olutur
    etkinlik_filtre = veri_gecici.etkinlik == e
    # veriyi etkinlie gre filtrele
    veri_filtreli = veri_gecici[etkinlik_filtre]
    
    # boy ve kilo iin etkinlik zelinde ortalama bul
    for s in boy_kilo_liste:
        ortalama = np.round(np.mean(veri_filtreli[s]),2)
        if ~np.isnan(ortalama): # eer etkinlik zelinde ortalama varsa
            veri_filtreli[s] = veri_filtreli[s].fillna(ortalama)
        else: # etkinlik zelinde ortalama yoksa tm veri iin ortalama bul
            tum_veri_ortalamasi = np.round(np.mean(veri[s]),2)
            veri_filtreli[s] = veri_filtreli[s].fillna(tum_veri_ortalamasi)
    # etkinlik zelinde kayp deerleri doldurulmu veriyi veri geiciye eitle            
    veri_gecici[etkinlik_filtre] = veri_filtreli

# kayp deerleri giderilmi geici veriyi gerek veri deikenine eitle
veri = veri_gecici.copy() 
veri.info() # boy ve kilo stunlarnda kayp deer saysna bak
```

## Anomali Tespiti
* Aykr deer, dier gzlemlerden nemli lde farkl bir veri noktasdr. Bir aykr deer, lmdeki deikenlie bal olabilir veya deneysel hatay gsterebilir; ikincisi bazen veri kmesinden karlr. Bir aykr deer istatistiksel analizlerde ciddi sorunlara neden olabilir.
* 

```python
# buraya kod yazacaz
def anomaliTespiti(df,ozellik):
    outlier_indices = []
    
    for c in ozellik:
        # 1. eyrek
        Q1 = np.percentile(df[c],25)
        # 3. eyrek
        Q3 = np.percentile(df[c],75)
        # IQR: eyrekler akl
        IQR = Q3 - Q1
        # aykr tespiti iin arpan
        outlier_step = IQR * 1.5
        # aykry ve aykr indeksini tespit et
        outlier_list_col = df[(df[c] < Q1 - outlier_step) | (df[c] > Q3 + outlier_step)].index
        # indeksleri depola
        outlier_indices.extend(outlier_list_col)
        
    # esiz aykr deerleri bul
    outlier_indices = Counter(outlier_indices)
    # eer bir rnek (v) 1 farkl stun iin aykr deerse bunu aykr olarak kabul et (v>1)
    multiple_outliers = list(i for i, v in outlier_indices.items() if v > 1)
    
    return multiple_outliers
```

```python
# buraya kod yazacaz
veri_anomali = veri.loc[anomaliTespiti(veri, ["yas", "kilo", "boy"])]
veri_anomali.spor.value_counts()
```
