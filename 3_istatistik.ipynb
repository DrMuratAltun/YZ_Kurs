### Merkezi Eilim

* Mode: Bir listede en ok bulunan saydr.
* Ortalama ve Medyan:

```python
# buraya kod yazacaz
yas = [1,2,3,5,6,7,7,10,12,13]
# ortalama 
ortalama_yas = np.mean(yas)
print("Ortalama: ", ortalama_yas)

# ortanca ya da medyan 
medyan_yas = np.median(yas)
print("Medyan: ", medyan_yas)

# mod 
mod_yas = stats.mode(yas)
print("Mod: ", mod_yas)

```

    * Varyans deiikliktir. Listedeki tm noktalarn ortalamadan ne kadar uzaklkta olduuna ilikin bir lttr.

```python
# buraya kod yazacaz
print("Varyans: ", (np.var(yas)))
var = sum( (yas - np.mean(yas))**2 / len(yas) )
print("Yukardaki forml ile kendi hesapladmz varyans:", var)

print("Standart Sapma: ", (np.std(yas)))
std = np.sqrt(sum( (yas - np.mean(yas))**2 / len(yas) )) 
print("Yukardaki forml ile kendi hesapladmz standart sapma:", std)
```

### Drttebirlik
* Drttebirlik nedir?
    * Drttebirlik sralanm bir veri setini drt eit paraya blen ve bylece her bir blnen parann ana ktle veya rneklem verilerinin 1/4'ini kapsad,  tane zetleme deeridir
    * 1,4,5,6,8,9,11,12,13,14,15,16,17
    * Bu listenin medyan deeri, listenin ortasnda bulunan say olan 11'dir.
    * Alt drttebir (ilk eyreklik (Q1)(25%)) en kk say ile medyan arasnda kalan listenin medyan yani 6'dr.
    * st drttebir (nc eyreklik (Q3)(75%)), en byk say ile medyan arasnda kalan listenin medyan yani 14'dr.
* eyrekler akl (Inter Quartile Range - IQR) = Q3-Q1
* Aykr deerler: Q1 - 1.5*IQR ve Q3 + 1.5*IQR snrlar darsnda kalan deerlerdir.
* Drttebirlik, eyrek olarak da isimlendirilir.

```python
# buraya kod yazacaz
veri = pd.read_csv("istatistik.csv")
veri.head()
```

### Pearson Korelasyon Katsays
* Pearson Korelasyon Katsays: Kovaryans, deikenlerin standart sapmasna blerek elde edilir.
* Yar ap ve alan arasndaki Pearson korelasyon katsaysna bakalm
* Pearson korelasyon katsays, +1 ve -1 deerleri arasnda deiir.
    * +1 = pozitif dorusal korelasyon
    * -1 = negatif dorusal korelasyon

* 

```python
# buraya kod yazacaz
p1 = veri.loc[:,["alan", "yari_cap"]].corr(method = "pearson")
p2 = veri.yari_cap.cov(veri.alan) / (veri.yari_cap.std() * veri.alan.std())

print("Pearson Korelasyon Katsays: ")
print(p1)
print("Pearson Korelasyon Katsays: ", p2)
```

### Etki Bykl
* zet istatistiklerden biridir.
* Bir etki boyutunu tanmlar. ki grup arasndaki fark lmenin basit bir yoludur.
* Baka bir deyile, etki bykl, farkn boyutunu vurgular.
* Cohen, d (etki boyutu) = 0.2 ise kk etki boyutu, d = 0.5 orta etki boyutu, d = 0.8 byk etki boyutu olduunu ne srmektedir.
* yi huylu ve kt huylu kanser hcrelerinin yar aplarnn etki bykln karlatralm
* Etki bykl forml
    * 

```python
# buraya kod yazacaz
# veriyi iki gruba ayralm
veri_iyi = veri[veri["teshis"] == "iyi_huylu"]
veri_kotu = veri[veri["teshis"] == "kotu_huylu"]

# yukardaki denklemdeki pay ifadesi
# M1: kt huylu hcrelerin yarap ortalamas
# M2: iyi huylu hcrelerin yarap ortalamas
pay = veri_kotu.yari_cap.mean() - veri_iyi.yari_cap.mean()

# yukardaki denklemdeki payda ifadesindeki S deerleri
# S1^2 : kt huylu hcrelerin yarap varyans
# S2^2 : iyi huylu hcrelerin yarap varyans
iyi_varyans = veri_iyi.yari_cap.var()
kotu_varyans = veri_kotu.yari_cap.var()

# n1 : kt huylu hcrelere ilikin veri says
# n2 : iyi huylu hcrelere ilikin veri says
n1 = len(veri_kotu)
n2 = len(veri_iyi)

payda = np.sqrt(((n2*kotu_varyans) + (n1*iyi_varyans)) / float(n1 * n2))

effect_size = pay / payda
print("Etki Bykl: ", effect_size)





```

* Etki bykl 21'dir ve beklediimiz gibi iki grubun birbirinden farkl olduunu sylyor. nk gruplarmz iyi huylu kanser hcresi yarap ve kt huylu kanser hcresi yarap, birbirinden farkl anlamlara geliyor.

### Permtasyon
* Matematikte permtasyon, her semboln sadece bir veya birka kez kullanld sral bir dizidir.
* Mesela 3 tane renci olsun: Ali, Veli ve Aye
* Bu  renciyi ka farkl ekilde yan yana sralarz? 3! = 3.2.1 = 6
    * Ali, Veli, Aye
    * Ali, Aye, Veli
    * Veli, Aye, Ali
    * Veli, Ali, Aye
    * Aye, Veli, Ali
    * Aye, Ali, Veli
* n elemanlk bir kmenin permtasyonu n!
* n elemanlk bir kmenin r alana permtasyonu ise (tekrarlamak yasak)
* 
* rnek:
    * Bir web sitesine ye olacaksnz; 5 karakterli bir ifre belirleyeceksiniz
    * Karakterler kk harf ya da rakamlar (0-9 aras saylar) olabilir
    * Kullanlan bir karakter ya da rakam tekrar kullanmak yok
    * Ka farkl ifre yaratabilirsiniz?
        * n = 29(harf says) + 10(rakam says) = 39
        * r = 5
        * P = 39! / (39-5)! = 39! / (34!)

```python
# buraya kod yazacaz
import math
pay = math.factorial(39)
payda = math.factorial(34)
P = int(pay / payda)
print(P)
```

## Bayes Teoremi

* P(A|B) = B olay gerekletiinde A olaynn gerekleme olasl
* P(A) = A olaynn gerekleme olasl
* P(B|A) = A olay gerekletiinde B olaynn gerekleme olasl
* P(B) = B olaynn gerekleme olasl
* imdi bu formuln nasl ktna bakalm
    * P(A|B) = P(A ve B) / P(B)
    * P(B|A) = P(B ve A) / P(A)
    * P(A ve B) = P(B ve A)
    * P(A|B).P(B)=P(B|A).P(A)
    * P(A|B) = P(B|A) . P(A) / P(B)
* rnek:
    * Bir okul var. Bu okulda 50 tane ocuk olsun. Bu 50 ocuktan 1 tanesi x hastalna yakalanabilir. x hastalna sahip hastalkl bir ocuun testi %90 pozitif, salkl bir ocuun testi ise %10 pozitif sonu verebilir. Test sonucu pozitif olan bir ocuun gerekten x hastalna sahip olma olasl nedir?
       * P(A) : ocuun hasta olmas olasl = 1/50
       * P(B) : Testin pozitif kmas olasl = 1/50 * 0.90 + 49/50 * 0.10 = 0.116
       * P(A|B) : Pozitif kan testin hastalk kma olasl ? 
       * P(B|A) : x hastalna sahip ocuun testinin pozitif kma olasl = 0.9
       * P(A|B)=P(B|A)*P(A)/P(B) => (0.9 * 1/50) / (0.116) = %15
